```python
import requests as req
from bs4 import BeautifulSoup as bs
from datetime import datetime
import pandas as pd
import csv
import os
hades = {'user-agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/106.0.0.0 Safari/537.36'}
```

```python
def scraping_liputan6():
    a = 1
    data_list = []
    for page in range(3):  # Sesuaikan berapa halaman yang ingin discraping
        url = f'https://www.liputan6.com/'


        halaman_utama = req.get(url, headers=hades).text
        memanggil_beautifulsoup = bs(halaman_utama, 'lxml')
        div_utama = memanggil_beautifulsoup.find('div', class_='container-main')
        mencari_semua_artikel = div_utama.find_all('article', class_='articles--iridescent-list--text-item')

        for x in mencari_semua_artikel:
            judul = x.find('a', class_='ui--a articles--iridescent-list--text-item__title-link').text
            link = x.find('a')['href']
            highlight = x.find('div', class_='articles--iridescent-list--text-item__summary articles--iridescent-list--text-item__summary-seo').text
            kategori_element = x.find('a', class_='articles--iridescent-list--text-item__category')
            kategori = kategori_element.text.strip() if kategori_element else 'Kategori tidak ditemukan'
            tanggal_elemen = x.find('span', class_='articles--iridescent-list--text-item__datetime')

            if tanggal_elemen:
                time_elemen = tanggal_elemen.find('time')
                if time_elemen and time_elemen.has_attr('datetime'):
                    datetime_str = time_elemen['datetime']
                    dt = datetime.fromisoformat(datetime_str)
                    formatted_date = dt.strftime('%d %b %Y, %H:%M WIB')
                else:
                    formatted_date = 'Tidak dicantumkan'
            else:
                formatted_date = 'Tidak dicantumkan'


            halaman_artikel = req.get(link, headers=hades).text
            memanggil_beautifulsoup_isiartikel= bs(halaman_artikel, 'lxml')

            penulis = memanggil_beautifulsoup_isiartikel.find('span', class_='read-page--header--author__name fn')
            nama_penulis = penulis.text.strip() if penulis else 'Penulis tidak ditemukan'

            print(f'\033[1mBerita ke {a}\033[0m \nJudul: {judul} \nKatagori: {kategori} \nPenulis: {nama_penulis} \nTanggal Upload :{formatted_date} \nHighlight Berita: {highlight} \nLink: {link} \n---------------------------------------------------------------------------------------------------------------------------------------------------------')
            a += 1

            data_list.append([judul, kategori, nama_penulis, formatted_date, highlight, link])



            # Menyimpan hasil scraping ke dalam file CSV
            filename = 'Hasil_Scraping_Berita_Liputan6.csv'
            file_exists = os.path.isfile(filename)
            with open(filename, 'a', newline='', encoding='utf-8') as file:
                wr = csv.writer(file, delimiter=',')
                if not file_exists:
                    wr.writerow(['Judul', 'Kategori', 'Author', 'Tanggal', 'Highlight', 'Link'])
                wr.writerow([judul, kategori, nama_penulis, formatted_date, highlight, link])

    # Menyimpan hasil scraping ke dalam file XLSX
    df = pd.DataFrame(data_list, columns=['Judul', 'Kategori', 'Author', 'Tanggal', 'Highlight', 'Link'])
    filename_xlsx = 'Hasil_Scraping_Berita_Liputan6.xlsx'
    df.to_excel(filename_xlsx, index=False, engine='openpyxl')
```

```python
scraping_liputan6()
```
